# ML on Iris flower data set: (naive_bayes)

## load data set:

```python
from sklearn.datasets import load_iris

iris = load_iris()
iris
```

## make data frame: 

```python
import pandas as pd
import numpy as np

print("all keys: ", iris.keys())
print("feature name:", iris.feature_names)
print("target column", iris.target)
print("target name", iris.target_names)

data = iris.data
label = iris.feature_names

df = pd.DataFrame(data=data)
df.columns = label
df['type'] = iris.target
df
```

## make x and y:

```python
x = df[["sepal length (cm)","sepal width (cm)", "petal length (cm)","petal width (cm)"]]
y = df[['type']]
```

## split the train and test:

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=0.2, random_state=3)
```

## make and fit data on model: (DecisionTreeClassifier)

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(criterion='entropy', max_depth=5, min_samples_split=7, min_samples_leaf=2, class_weight={0: 0.33, 1: 0.33, 2: 0.33}, random_state=2)
model.fit(train_x, train_y)
pred = model.predict(test_x)
```

## model score in GaussianNB model:

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print("confusion_matrix: \n",confusion_matrix(y_true=test_y, y_pred=pred))
print("accuracy_score: ",accuracy_score(y_true=test_y, y_pred=pred))
print(classification_report(y_true=test_y, y_pred=pred))
```

## plot tree:

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt

plt.figure(figsize=(12,8))
plot_tree(model, filled=True, feature_names=iris.feature_names, class_names=iris.target_names)
plt.show()
```
